## Question Six: What are the peak hours of sales?

```python
print(df7['time'].dtype)
```

```python
## Convert to datetime
df7['time']=pd.to_datetime(df7['time'], format='%H:%M:%S', errors='coerce')
df7.head()
```

```python
## Let us extract hour, month and day
df7['hour']=df7['time'].dt.hour
df7['Year']=df7['date'].dt.year
df7['Month']=df7['date'].dt.month_name().str[:3]
df7['Day']=df7['date'].dt.day_name().str[:3]
```

```python
df7.head()
```

```python
## Group by hour to find peak hours of sales
peak_hours =df7.groupby('hour')['quantity'].sum().sort_values(ascending=False)
peak_hours
```

```python

# Plotting as a bar chart
plt.bar(peak_hours.index, peak_hours.values, color="navy")
plt.title("Peak Hours of Sales Count")
plt.xlabel("Hour of the Day")
plt.ylabel("Quantity Sold")
plt.xticks(peak_hours.index)  # Ensure all hour labels are shown
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Question 7: Find the total sales made on each day of the week. Which day of the week is when sales are made the most?


```python
Peak_week_day = df7.groupby('Day')['price'].sum().sort_values(ascending=False)
Peak_week_day
```

```python
# PLotting as a bar chart of the week when sales are made the most?
plt.bar(Peak_week_day.index, Peak_week_day.values, color="brown")
plt.title("Peak Week Day")
plt.xlabel("Week Days")
plt.ylabel("Sum of Sales per Weekday")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Question 8: Find the top 5 bestselling pizzas.


```python
top_5_pizzas= df7.groupby('pizza_type_id')['quantity'].sum().sort_values(ascending=False).head(5)
top_5_pizzas
```

```python
plt.figure(figsize=(7,7))
plt.pie(top_5_pizzas.values, labels=top_5_pizzas.index, autopct='%1.1f%%', startangle=140)
plt.title("Top 5 Bestselling Pizzas")
plt.axis('equal')  # Ensures the pie is drawn as a circle.
plt.show()
```

## Question 9:  Find the sales made in each month. Any trend noticeable?

```python
monthly_sales=df7.groupby('Month')['price'].sum().sort_values(ascending=False)

# convert to DataFrame an reset index
monthly_sales_df = monthly_sales.reset_index()

#Renames columns
monthly_sales_df.columns = ['Month', 'Monthly Sales']

#Display
print("Sales made in each month are:", monthly_sales_df)
```
